Leitura e conversão dos Datasets em formato csv pra DataFrame do pandas

```python
datasets_info = {
    "df_customers": "Datasets/olist_customers_dataset.csv",
    "df_geolocation": "Datasets/olist_geolocation_dataset.csv",
    "df_order_items": "Datasets/olist_order_items_dataset.csv",
    "df_order_payments": "Datasets/olist_order_payments_dataset.csv",
    "df_orders_reviews": "Datasets/olist_order_reviews_dataset.csv",
    "df_orders": "Datasets/olist_orders_dataset.csv",
    "df_products": "Datasets/olist_products_dataset.csv",
    "df_sellers": "Datasets/olist_sellers_dataset.csv",
    "df_product_name_translation": "Datasets/product_category_name_translation.csv"
}

dataframes = {}

for var_name, file_path in datasets_info.items():
    reader = cls.CsvToDataFrame(file_path)
    dataframes[var_name] = reader.read_csv()

df_customers = dataframes["df_customers"]
df_geolocation = dataframes["df_geolocation"]
df_order_items = dataframes["df_order_items"]
df_order_payments = dataframes["df_order_payments"]
df_orders_reviews = dataframes["df_orders_reviews"]
df_orders = dataframes["df_orders"]
df_products = dataframes["df_products"]
df_sellers = dataframes["df_sellers"]
df_product_name_translation = dataframes["df_product_name_translation"]y
```

Cria a pasta staging_area, aplica o tratamento dos valores nulos e duplicados, salva na staging_area e exibe a mensagem indicando se o arquivo foi salvo.

```python

os.makedirs("staging_area", exist_ok=True)

datasets = [
    (df_customers, "dataset_customers.csv"),
    (df_geolocation, "dataset_geolocation.csv"),
    (df_order_items, "dataset_order_items.csv"),
    (df_order_payments, "dataset_order_payments.csv"),
    (df_orders_reviews, "dataset_orders_reviews.csv"),
    (df_orders, "dataset_orders.csv"),
    (df_products, "dataset_products.csv"),
    (df_sellers, "dataset_sellers.csv"),
    (df_product_name_translation, "dataset_product_name_translation.csv")
]

for df, nome_arquivo in datasets:
    cleaner = cls.CleaningDatasets(df)
    df_no_nulls = cleaner.fill_and_report_nulls()
    df_clean = cleaner.remove_duplicate(df_no_nulls)
    df_clean.to_csv(f"staging_area/{nome_arquivo}", index=False)
    print(f"Arquivo salvo: staging_area/{nome_arquivo}")
```
